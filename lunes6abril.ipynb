```python
#ejercicios
a= 10
print(a)

b=10.0
print(b)

c= """
hola"""
print(c)

d= c*a
print(d)

e= True
(e and e)
(e and not e)

l = ["a","b",1,True] #mixed type list
l.append("hola")
l.pop(1) #pop es pa quitar,se elimina el b

#listas por comprensión
comp_list = [i for i in range(10) if i%2==0] #esto es una lista por comprensión, se crea una lista con los números del 0 al 9 que sean pares
print(comp_list)

#tuplas
t= (1,2,3) #las tuplas son inmutables, no se pueden modificar

#diccionarios
d = {"eduardo": {"edad": 26, "ciudad": "Madrid"}, "diego": {"edad": 30, "ciudad": "Barcelona"}}#los diccionarios son mutables, se pueden modificar
print(d["eduardo"]["edad"])
print(d["diego"]["edad"])
```

## Condicionales

```python
if d["eduardo"]["edad"] > 18:
    print("Eduardo es mayor de edad")
elif d["eduardo"]["edad"]< 0:
    print("Eduardo no ha nacido")
else:    print("Eduardo es menor de edad")
```

## Excepciones : controlar muchas variables

```python
try:  #intentar hacer algo, si no se puede, ir al except
    print(d["eduardo"]["altura"]) #esto va a dar error porque no existe la clave "nombre" en el diccionario de eduardo
except Exception as e: #capturar el error y guardarlo en la variable e
    print("Error:", e) #imprimir el error
```

## Número aleatorios

```python
#si lo ejecuto varias veces, me va a dar un número diferente cada vez, porque es aleatorio
np.random.rand(1,10)
```

```python
#pa fijar el número aleatorio, para que siempre me de el mismo resultado, se usa la función seed
np.random.seed(42)
np.random.rand(1,10)
```

```python
random_array = np.random.rand(1,10) #crear un array de 1x10 con números aleatorios entre 0 y 1
random_array
```

```python
my_array = random_array.reshape(5,2) #cambiar la forma del array a 5x2 FILAS, COLUMNAS 
my_array
```

```python
#sacar la traspuesta de un array
my_array.T
```

```python
unos = np.ones((2,5)) #crear un array de 2x5 con unos
unos
```

```python
my_array2 = my_array.reshape(2,5) #deben estar ambas en la misma dimensión para poder sumarlas
unos * 2 + my_array2 #agregar ruido a mi array

my_array2.std() #desviación estándar
my_array2.std() ** 2 #varianza de mi array
```

## Resolver problemas de ecuaciones con matrices

```python
A= np.random.rand(3,3) #crear un array de 3x3 con números aleatorios entre 0 y 1
A
```

```python
B = np.random.rand(3,1) #crear un array de 3x1 con números aleatorios entre 0 y 1
B
```

```python
np.linalg.solve(A,B) #resolver el sistema de ecuaciones Ax=B, donde A es la matriz de coeficientes, B es el vector de términos independientes y x es el vector de incógnitas
```
